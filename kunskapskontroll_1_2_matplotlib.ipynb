### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
Explicit / Implicit. Recommended approach = Explicit.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure can be looked at as a paper, or a canvas. Axes are the seperate charts on the paper/canvas (There can be several axes on the same Figure), each with their
rules, lines etc.
Axis = objects that set the scale and limits, as well as marks the ticks with ticklabels etc.
Artist is everything that is visible on the paper/canvas (Figure).
```

### When plotting in Matplotlib, what is the expected input data type?

```python
Numpy array, or objects that can be passed to numpy arrays (like pandas, lists, tuples etc.)
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 10)
y = x**2

#object oriented
fig, ax = plt.subplots()

ax.set_title("Plot of the function y = x^2 (from -4 to 4)")
ax.set_xlabel("x")
ax.set_ylabel("y")

ax.plot(x, y)

#pyplot
plt.title("Plot of the function y = x^2 (from -4 to 4)")
plt.xlabel("x")
plt.ylabel("y")

plt.plot(x, y)

```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, axs = plt.subplots(2, 1)

axs[0].scatter(random_data_x, random_data_y)
axs[0].set_title("Scatter plot")
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")

axs[1].bar(names, values)
axs[1].set_title("Bar plot")
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
```
